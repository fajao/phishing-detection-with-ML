```python
######################################################################
# Checking how many malicious URLs are reachables via whois command. #
######################################################################
import pandas as pd
import numpy as np
import whois
import concurrent.futures
import time
```

```python
# Read the csv file
urls = pd.read_csv('final_raw_data.csv', encoding = "ISO-8859-1")
```

```python
# Whois Function
def performwhois(url):
    try:
        result = whois.whois(url)
        return True #success
    except Exception as e:
        print(f"WHOIS lookup failed for {url}: {e}")
        return False #error
```

```python
# Sample list of URLs to process
sample_1 = urls['URL'][49524:50150]
sample_2 = urls['URL'][1289000:1298453]

phish_sample = pd.concat([sample_1, sample_2])
```

```python
phish_urls = []
counter = 0
for url in phish_sample:
    if performwhois(url):
        counter = counter + 1
        print(counter)
        phish_urls.append(url)
```
